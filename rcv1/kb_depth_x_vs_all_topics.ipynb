```python
import matplotlib.pyplot as plt

train_sizes = [12, 60, 120, 600, 1200]
depth_0_micro_f1 = [0.7009, 0.8912, 0.9099, 0.9233, 0.9284]
depth_1_micro_f1 = [0.6947, 0.8654, 0.8895, 0.9209, 0.9301]
depth_2_micro_f1 = [0.6782, 0.8188, 0.8655, 0.9060, 0.9189]
all_topics_micro_f1 = [0.6413, 0.8125, 0.8546, 0.8765, 0.8974]

plt.title('Micro $F_1$ Score vs Training Set Size on RCV1')
plt.ylabel('Micro $F_1$ Score')
plt.xlabel('Training Set Size')

plt.plot(train_sizes, depth_0_micro_f1, '-^', label='Knowledge Base Depth 0')
plt.plot(train_sizes, depth_1_micro_f1, '-s', label='Knowledge Base Depth 1')
plt.plot(train_sizes, depth_2_micro_f1, '-D', label='Knowledge Base Depth 2')
plt.plot(train_sizes, all_topics_micro_f1, '-o', label='Knowledge Base All Topics')
plt.xticks([12, 200, 400, 600, 800, 1000, 1200])
plt.legend(loc='lower right')
plt.savefig('depth_x_vs_all_topics_rcv1.pdf', bbox_inches='tight')
plt.show()
```
