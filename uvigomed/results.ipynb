# Plots of Classification Results

Load imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

Create arrays containing the data.

```python
training_set_sizes = [5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000, 43747]

# Multinomial Naive Bayes
nb_micro = [0.084350, 0.099778, 0.126571, 0.153640, 0.235218, 0.315133, 0.406781, 0.447524, 0.502033,
            0.563470, 0.597099, 0.625831, 0.649760]
nb_macro = [0.015355, 0.030799, 0.059436, 0.080097, 0.112873, 0.149942, 0.204450, 0.230261, 0.268867,
            0.318991, 0.354513, 0.389438, 0.430352]

# Support Vector Classifier
svc_micro = [0.082594, 0.091463, 0.126940, 0.162971, 0.250185, 0.338137, 0.460181, 0.519217, 0.583610,
             0.656319, 0.688747, 0.717295, 0.747783]
svc_macro = [0.012462, 0.024414, 0.051996, 0.092684, 0.137591, 0.204805, 0.299433, 0.350110, 0.415942,
             0.500415, 0.532174, 0.562759, 0.597271]

# Knowledge Base Embedding
kb_micro = [0.050288, 0.063402, 0.098028, 0.166053, 0.271629, 0.419096, 0.605812, 0.637985, 0.663553,
            0.682612, 0.696669, 0.713652, 0.729786]
kb_macro = [0.003683, 0.009576, 0.023147, 0.066125, 0.137898, 0.250478, 0.388701, 0.438939, 0.478753,
            0.511282, 0.531018, 0.565247, 0.586063]

# Bag of Concepts from:
# García, M.A.M., Rodríguez, R.P. and Rifón, L.E.A., 2015.
# Biomedical literature classification using encyclopedic knowledge: a wikipedia-based bag-of-concepts approach.
# PeerJ [Online]. Available from: https://doi.org/10.7717/peerj.1279.
boc_training_set_sizes = [5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000]
boc_macro = [0.017, 0.040, 0.078, 0.116, 0.179, 0.269, 0.331, 0.390, 0.430, 0.467]
```

```python
plt.title('Micro $F_1$ Score vs Training Set Size')
plt.ylabel('Micro $F_1$ Score')
plt.xlabel('Training Set Size')
plt.plot(training_set_sizes, nb_micro, '-.', label='Multinomial Naive Bayes')
plt.plot(training_set_sizes, svc_micro, '--', label='Support Vector Machine')
plt.plot(training_set_sizes, kb_micro, '-', label='Knowledge Base')
plt.legend(loc='lower right')
plt.savefig('micro_f1_large_uvigomed.pdf', bbox_inches='tight')
plt.show()
```

```python
plt.title('Micro $F_1$ Score vs Training Set Size')
plt.ylabel('Micro $F_1$ Score')
plt.xlabel('Training Set Size')
plt.plot(training_set_sizes, nb_micro, '-o', label='Multinomial Naive Bayes')
plt.plot(training_set_sizes, svc_micro, '-s', label='Support Vector Machine')
plt.plot(training_set_sizes, kb_micro, '-D', label='Knowledge Base')
plt.xlim(100, 5000)
plt.xticks([100, 1000, 2000, 3000, 4000, 5000])
plt.legend(loc='lower right')
plt.savefig('micro_f1_small_uvigomed.pdf', bbox_inches='tight')
plt.show()
```

```python
plt.title('Macro $F_1$ Score vs Training Set Size')
plt.ylabel('Macro $F_1$ Score')
plt.xlabel('Training Set Size')
plt.plot(training_set_sizes, nb_macro, '-.', label='Multinomial Naive Bayes')
plt.plot(training_set_sizes, svc_macro, '--', label='Support Vector Machine')
plt.plot(training_set_sizes, kb_macro, '-', label='Knowledge Base')
plt.legend(loc='lower right')
plt.savefig('macro_f1_large_uvigomed.pdf', bbox_inches='tight')
plt.show()
```

```python
plt.title('Macro $F_1$ Score vs Training Set Size')
plt.ylabel('Macro $F_1$ Score')
plt.xlabel('Training Set Size')
plt.plot(training_set_sizes, nb_macro, '-o', label='Multinomial Naive Bayes')
plt.plot(training_set_sizes, svc_macro, '-s', label='Support Vector Machine')
plt.plot(training_set_sizes, kb_macro, '-D', label='Knowledge Base')
plt.plot(boc_training_set_sizes, boc_macro, '-^', label='Bag of Concepts')
plt.xlim(100, 5000)
plt.xticks([100, 1000, 2000, 3000, 4000, 5000])
plt.legend(loc='lower right')
plt.savefig('macro_f1_small_uvigomed.pdf', bbox_inches='tight')
plt.show()
```
